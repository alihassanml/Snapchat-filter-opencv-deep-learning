```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as imd
import tensorflow as tf
import cv2
```

```python
url1 = './img.jpg'
url2 = './man.jpeg'
```

```python
img = plt.imread(url1)
plt.imshow(img)
plt.axis('off')
```

```python
def detect_face(img):
    haarcascade_eye = './detect/haarcascade_eye.xml'
    eye_cased = cv2.CascadeClassifier(haarcascade_eye)
    eye = eye_cased.detectMultiScale(img)
    return eye        
```

```python
eye_x0, eye_y0, eye_w0, eye_h0 = eye[0]
eye_x1, eye_y1, eye_w1, eye_h1 = eye[1]
```

```python
x = min(eye_x0, eye_x1)
y = min(eye_y0, eye_y1)
w = (eye_x1 + eye_w1) - eye_x0  
h = max(eye_h0, eye_h1)
```

```python
img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.rectangle(img_1, (x, y), (x + w, y + h), (255, 0, 0), 2)
plt.imshow(img_1)
plt.show()
```

```python
eye = detect_face(plt.imread(url2))
eye_x0, eye_y0, eye_w0, eye_h0 = eye[0]
eye_x1, eye_y1, eye_w1, eye_h1 = eye[1]

# sum both eyes
x = min(eye_x0, eye_x1)
y = min(eye_y0, eye_y1)
w = (eye_x1 + eye_w1) - eye_x0  
h = max(eye_h0, eye_h1)
```

```python
img_1 = cv2.cvtColor(plt.imread(url2), cv2.COLOR_BGR2RGB)
img_1 = cv2.rectangle(img_1, (eye_x0, eye_y0), (eye_x0 + eye_w0, eye_y0 + eye_h0), (0,255,255), 2)
plt.imshow(img_1)
plt.show()
```
